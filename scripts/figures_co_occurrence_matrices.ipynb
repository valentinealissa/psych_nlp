```python
sociodemo_groups = [male_black, male_white]
sociodemo_groups_string = ['male_black', 'male_white']

# version 2: looks at comorbidities as unique conditions per patient with encounter level comorbidity
def co_occurrence_matrix(df, name):
    num_patient = df['person_id'].nunique()
    co_mat = pd.crosstab(df.etl_epic_encounter_key, df.CCSR_code)
    co_occ = co_mat.T.dot(co_mat)
    final_cocc = np.clip(co_occ, 0, 0)

    for patient_id, patient_data in df.groupby('person_id'):
        co_mat = pd.crosstab(patient_data.etl_epic_encounter_key, patient_data.CCSR_code)
        co_occ = co_mat.T.dot(co_mat)
        np.fill_diagonal(co_occ.values, 0)
        patient_cocc = np.clip(co_occ, None, 1)
        final_cocc, patient_cocc = final_cocc.align(patient_cocc, fill_value = 0)
        final_cocc = np.add(final_cocc, patient_cocc)
    # print(final_cocc)

    co_occ_norm = final_cocc.div(num_patient, fill_value=0)
    plot_name = "Co-occurrence matrix for {0}".format(name)
    figure_name = "../figures/co_occurrence_heatmaps/ccsr/norm/patient_level_v2/co_occ_norm_ccsr_{0}.png".format(name)
    sns.heatmap(co_occ_norm, cmap = "cubehelix_r") #, vmin=0, vmax=0.09)#,linewidth=.1) #cmap='cubehelix_r',
    plt.title(plot_name)
    plt.savefig(figure_name, transparent=False, format='png', dpi=300, bbox_inches='tight')
    plt.clf()

for i,group_df in enumerate(sociodemo_groups):
    co_occurrence_matrix(group_df, sociodemo_groups_string[i])
```
