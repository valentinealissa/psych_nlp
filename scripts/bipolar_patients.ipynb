```python
bipolar_patients_all_d = bipolar_patients.sort_values(by=['PrimaryMrn','VisitDate'], ascending=[True, True])
clean_bipolar_patients = bipolar_patients_all_d.drop_duplicates(subset=['PrimaryMrn', 'DisplayString'],  keep='first')
clean_bipolar_patients
```

```python
ccsr = pd.read_csv("/Users/valena17/DataGripProjects/msdw2/psych_diagnoses/ccsr_fcodes.csv")
ccsr = ccsr.applymap(lambda x: x.replace("'", ""))
ccsr = ccsr[['ICD-10-CM CODE', 'Default CCSR CATEGORY DESCRIPTION OP']]
ccsr_mapping = ccsr.set_index('ICD-10-CM CODE').T.to_dict('list')
ccsr_mapping['F01518']

def CCSR(row):
    row = row.replace('.', '')
    if row in ccsr_mapping.keys():
        return ccsr_mapping[row][0]

clean_bipolar_patients['CCSR_code'] = clean_bipolar_patients['Value'].apply(CCSR)
clean_bipolar_patients.head(50)
```

```python
clean_ccsr_bipolar_patients = clean_bipolar_patients.drop_duplicates(subset=['PrimaryMrn', 'CCSR_code'],  keep='first')
clean_ccsr_bipolar_patients
```

```python
def condition_type(row):
    if row == 'Bipolar and related disorders':
        return 1
    else:
        return 0

clean_ccsr_bipolar_patients['bp_condition'] = clean_ccsr_bipolar_patients.CCSR_code.apply(lambda row: condition_type(row))

clean_ccsr_bipolar_patients2 = clean_ccsr_bipolar_patients.groupby('PrimaryMrn').filter(lambda x: x['bp_condition'].mean() > 0)
```

```python
male = clean_ccsr_bipolar_patients2.loc[clean_ccsr_bipolar_patients2.Merged_Gender.isin(['Male'])]
male_asian = male.loc[male.Race.isin(['Asian'])]
male_black = male.loc[male.Race.isin(['Black or African American'])]
male_latino = male.loc[male.Race.isin(['Hispanic or Latino'])]
male_white = male.loc[male.Race.isin(['White'])]

female = clean_ccsr_bipolar_patients2.loc[clean_ccsr_bipolar_patients2.Merged_Gender.isin(['Female'])]
female_asian = female.loc[female.Race.isin(['Asian'])]
female_black = female.loc[female.Race.isin(['Black or African American'])]
female_latina = female.loc[female.Race.isin(['Hispanic or Latino'])]
female_white = female.loc[female.Race.isin(['White'])]

print(male_black['PrimaryMrn'].nunique(),
      male_white['PrimaryMrn'].nunique(),
      female_black['PrimaryMrn'].nunique(),
      female_white['PrimaryMrn'].nunique())
```
