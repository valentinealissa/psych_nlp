```python
import pandas as pd
import random

random.seed(41)

df = pd.read_csv("~/psych_nlp/sentiment-analysis/data/sentences_MD-labels.csv", header=0)

df['label'] = pd.Series(
    random.choices(['entailment', 'not_entailment'], weights=[1, 1], k=len(df)), 
    index=df.index
)

df
```

```python
positive_h = "The sentiment of this sentence is positive."
negative_h = "The sentiment of this sentence is negative."
neutral_h = "The sentiment of this sentence is neutral."

for index, row in df.iterrows():
    if row['label'] == 'entailment':
        if row['MD_label'] == 'positive':
            df.at[index, "hypothesis"] = positive_h
        if row['MD_label'] == 'negative':
            df.at[index, "hypothesis"] = negative_h
        if row['MD_label'] == 'neutral':
            df.at[index, "hypothesis"] = neutral_h

    if row['label'] == 'not_entailment':
        if row['MD_label'] == 'positive':
            df.at[index, "hypothesis"] = random.sample([negative_h, neutral_h], 1)
        if row['MD_label'] == 'negative':
            df.at[index, "hypothesis"] = random.sample([neutral_h, positive_h], 1)
        if row['MD_label'] == 'neutral':
            df.at[index, "hypothesis"] = random.sample([positive_h, negative_h], 1)[0]

df['hypothesis'] = df['hypothesis'].astype('string')

df
```

```python
df.to_csv("../data/MD-NLI.csv", index=False)
```
