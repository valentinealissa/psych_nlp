```python
import pandas as pd
import random

random.seed(42)

sentences = pd.read_csv('sentences_MD-labels.csv', header=0)
sentences['idx'] = sentences.index

p = list(sentences[sentences['MD_label'] == 'positive']['language'])
neg = list(sentences[sentences['MD_label'] == 'negative']['language'])
neu = list(sentences[sentences['MD_label'] == 'neutral']['language'])

sentences['context_label'] = pd.Series(
    random.choices(['negative', 'positive', 'neutral'], weights=[1, 1, 1], k=len(sentences)), 
    index=sentences.index
)

MD_prompt = "If you are the provider, what is your attitude towards the patient?"
prompt = "What is the sentiment of the sentence?"

for index, row in sentences.iterrows():
    #print(index, row['language'])
    if row['context_label'] == 'positive':
        context_sentence = random.sample(p, 1)
        context_sentence = f"Input: {context_sentence[0]}. Prompt: {prompt} Output: {row['context_label']}."
        #print(context_sentence)
        sentences.at[index, "context"] = context_sentence

    if row['context_label'] == 'negative':
        context_sentence = random.sample(neg, 1)
        context_sentence = f"Input: {context_sentence[0]}. Prompt: {prompt} Output: {row['context_label']}."
        #print(context_sentence)
        sentences.at[index, "context"] = context_sentence

    if row['context_label'] == 'neutral':
        context_sentence = random.sample(neu, 1)
        context_sentence = f"Input: {context_sentence[0]}. Prompt: {prompt} Output: {row['context_label']}."
        #print(context_sentence)
        sentences.at[index, "context"] = context_sentence

sentences.to_csv("sentences_MD_wContext.csv", index=False)
```
