```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
from datasets import DatasetDict, Dataset
import pandas as pd

df = pd.read_csv('../data/13_words_39_sentences.csv', header=0)
sequences_to_classify = df[['language']]
sequences_to_classify = list(sequences_to_classify['language'])

# dataset = Dataset.from_pandas(sentences).rename_columns({'language': 'sentence', "MD_label": 'sentiment'})
# dataset = dataset.add_column('label', create_labels(dataset['sentiment']))
# train_test = dataset.train_test_split(0.35, seed = seed)
# dev_test = train_test['test'].train_test_split(0.5, seed = seed)
# raw_dataset = DatasetDict({
#     'train': train_test['train'],
#     'validation': dev_test['train'],
#     'test': dev_test['test']})
```

```python
candidate_labels = ['neutral', 'negative', 'positive']

output = classifier(sequences_to_classify, candidate_labels)

# for s in sequences_to_classify:
#     print(classifier(s, candidate_labels))
```

```python
# print(output)
final_label = []
for result in output:
        final_label.append(result["labels"][0])
```

```python
df["NLI_label"] = final_label
```

```python
df.to_csv("../data/NLI_labels.csv", index = False)
```
